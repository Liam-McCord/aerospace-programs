```python
%matplotlib widget
import csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataframe_plot = pd.read_csv("output.csv")
dataframe_array = np.array(dataframe_plot).T

# Graphing section
fig = plt.figure(figsize = (7,7))

ax = fig.add_subplot(111, projection='3d')
ax.set_box_aspect([1,1,1])

print(dataframe_array[0][1])

for n in range(2):
    m = n * 3
    ax.plot3D(dataframe_array[m + 0], dataframe_array[m + 1], dataframe_array[m + 2])
#for i in range(0,4):
#    ax.plot3D(dataframe_plot[dataframe_plot.columns[i]].values().tolist(), dataframe_plot[dataframe_plot.columns[i + 1]].values().tolist(), dataframe_plot[dataframe_plot.columns[i + 2]].values().tolist())
graphcounter = 0

leg = plt.legend(loc='upper left')

    
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

ax.set_xlabel('X')

ax.set_ylabel('Y')

ax.set_zlabel('Z')


x = 637100 * np.outer(np.cos(u), np.sin(v))
y = 637100 * np.outer(np.sin(u), np.sin(v))
z = 637100 * np.outer(np.ones(np.size(u)), np.cos(v))

ax.plot_surface(x, y, z, rstride=4, cstride=4, color='b')

# The area that the graph will show upon start
GraphAreaStart = 1000000000

ax.set_xlim3d(-1 * GraphAreaStart, GraphAreaStart)
ax.set_ylim3d(-1 * GraphAreaStart, GraphAreaStart)
ax.set_zlim3d(-1 * GraphAreaStart, GraphAreaStart)

plt.show()
dataframe_array
```
